```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/IMDB Dataset.csv')
df.head(5)
```

```python
import preprocessing
```

```python
df['review'] = df['review'].apply(preprocessing.remove_punc)
df.head(5)
```

```python
df['review'] = df['review'].apply(preprocessing.lowercase)
df.head(5)
```

```python
df['review'] = df['review'].apply(preprocessing.remove_whitespace)
df.head(5)
```

```python
df['review'] = df['review'].apply(preprocessing.remove_special_characters)
df.head(5)
```

```python
df['review'] = df['review'].apply(preprocessing.remove_stopwords)
df.head(5)
```

```python
df['review'] = df['review'].apply(preprocessing.stemmer)
df.head(5)
```

```python
#Saving to a csv file
df.to_csv('/content/Clean_data.csv',index = False)
```
